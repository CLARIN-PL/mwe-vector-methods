```python
import os

import conllu
import pandas as pd
```

```python
dir_path = 'PARSEME_1.2_Polish_Dataset/PL'
filename = 'train.cupt'

filepath = os.path.join(dir_path, filename)
```

```python
with open(filepath, "r", encoding="utf-8") as f:
    sentence = next(conllu.parse_incr(f))
    tok = sentence[0]
    column_names = ['sent_id'] + list(tok.keys())
    
df = pd.DataFrame(columns = column_names)
    
with open(filepath, "r", encoding="utf-8") as f:
    sent_id = 0
    for sentence in conllu.parse_incr(f):
        sentence_rows = [{'sent_id': sent_id,
                          'id' : tok['id'],
                          'form': tok['form'],
                          'lemma': tok['lemma'],
                          'upos': tok['upos'],
                          'xpos': tok['xpos'],
                          'feats': tok['feats'],
                          'head': tok['head'],
                          'deprel': tok['deprel'],
                          'deps': tok['deps'],
                          'misc': tok['misc'],
                          'parseme:mwe': tok['parseme:mwe']} for tok in sentence]

        df = df.append(sentence_rows, ignore_index=True)

        sent_id += 1

df
```

```python
# save dataframe to csv
df.to_csv(os.path.join(dir_path, 'train_df.tsv'), sep='\t', index=False)
```

```python
# read test dataframe from TSV file
df_test = pd.read_csv(os.path.join(dir_path, 'text_df.tsv'), sep='\t')

df_test
```

```python
# get number of potential incorrect MWEs
# df_non_mwe = df_test[df_test['parseme:mwe'] == '*']

correct_pos = ['ADJ', 'NOUN', 'VERB']

mwe_count = 0

for sent_id in df_test['sent_id'].unique():
    pos_list = df_test[df_test['sent_id'] == sent_id]['upos'].tolist()
    mwe_tag_list = df_test[df_test['sent_id'] == sent_id]['parseme:mwe'].tolist()
    
    for pos_ind, pos in enumerate(pos_list[:-1]):
        if pos in correct_pos and mwe_tag_list[pos_ind] == '*' and pos_list[pos_ind + 1] in correct_pos and mwe_tag_list[pos_ind + 1 ] == '*':
            mwe_count += 1
            
mwe_count
```
