```python
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report
```

```python
results = pd.read_csv('merged_predictions.tsv', sep='\t')
results = results.rename(columns={'mwe': 'Lemma'})
correct_mwe = pd.read_csv('correct_mwe.tsv', sep='\t')
incorrect_mwe = pd.read_csv('incorrect_MWE_kompozycyjne_polaczenia_plWN.csv', error_bad_lines=False)
results
```

```python
results_corr = results[results['is_correct'] == 1.0]
results_corr = pd.merge(results_corr, correct_mwe[["Lemma", "Part of Speech"]], on="Lemma", how="left")

# results_corr = results.rename(columns={'Lemma': 'lemma', 'Part of Speech': 'pos'})

results_verb_corr = results_corr[results_corr['Part of Speech'] == 'verb']

results_verb_corr
```

```python
results_inc = results[results['is_correct'] == 0.0]
results_inc = results_inc.rename(columns={'Lemma': 'lemma'})
results_inc= pd.merge(results_inc, incorrect_mwe[["lemma", "pos"]], on="lemma", how="left")

results_verb_incorr = results_inc[results_inc['pos'] == '1']

results_verb_incorr = results_verb_incorr.rename(columns={'lemma': 'Lemma', 'pos': 'Part of Speech'})

results_verb_incorr
```

```python
results_verb = results_verb_corr.append(results_verb_incorr)

results_verb
```

```python
verb_results_df = pd.DataFrame(columns=['Model', 'Macro F1-score', 'F1 Incorr', 'F1 Corr'])

for model_name in results_verb.columns.tolist()[4:10]:
    y_true = results_verb['is_correct'].tolist()
    y_pred = results_verb[model_name].tolist()
    
    evaluation_results = classification_report(y_true, y_pred, output_dict=True)
    
    f1_macro = evaluation_results['macro avg']['f1-score']
    f1_incorr = evaluation_results['0.0']['f1-score']
    f1_corr = evaluation_results['1.0']['f1-score']
    
    verb_results_df = verb_results_df.append({'Model': model_name, 'Macro F1-score': f1_macro, 'F1 Incorr': f1_incorr, 'F1 Corr': f1_corr}, ignore_index=True)
    
    # print(f'Results for {model_name}',
    #       f'Macro F1-score: {f1_macro}',
    #       f'Incorrect F1-score: {f1_incorr}',
    #       f'Correct F1-score: {f1_corr}',
    #       '\n',
    #      sep='\n')
    
verb_results_df
```
