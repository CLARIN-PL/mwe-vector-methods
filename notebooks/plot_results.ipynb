```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# set seaborn theme
sns.set_theme(style="whitegrid")
```

<h4>Bar plot for gains between Contextual and Non-Contextual Embeddings Macro F1-score</h4>

```python
model_names = ['LR', 'RF', 'CNN']

non_contextual_macro_f1_results = [0.81, 0.74, 0.39]
contextual_macro_f1_results = [0.91, 0.83, 0.47]

macro_f1_gains = [elem_list_2 - elem_list_1 for (elem_list_1, elem_list_2) in zip(non_contextual_macro_f1_results, contextual_macro_f1_results)]

ax = sns.barplot(x=model_names, y=macro_f1_gains, palette="muted")

fontsize = 14

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points',
                fontsize=fontsize)

ax.set_title('Gain between Contextual and \nNon-Contextual Embeddings', fontsize=fontsize)

ax.set_ylabel('Macro F1-score Gain', fontsize=fontsize)
ax.set_xlabel('Model', fontsize=fontsize)

plt.yticks([0.01 * i for i in range(0, 17, 2)], fontsize=fontsize)
plt.xticks(fontsize=fontsize)

ax.set(ylim=(0, 0.11))

plt.tight_layout()

plt.savefig('gain_macro_f1.pdf', dpi=600)
```

<h4>Bar plot for gains between Contextual and Non-Contextual Embeddings F1-score for incorrect MWE class</h4>

```python
model_names = ['LR', 'RF', 'CNN']

non_contextual_macro_f1_results = [0.83, 0.81, 0.78]
contextual_macro_f1_results = [0.98, 0.97, 0.94]

macro_f1_gains = [elem_list_2 - elem_list_1 for (elem_list_1, elem_list_2) in zip(non_contextual_macro_f1_results, contextual_macro_f1_results)]

ax = sns.barplot(x=model_names, y=macro_f1_gains, palette="muted")

fontsize = 14

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points',
                fontsize=fontsize)

ax.set_title('Gain between Contextual and Non-Contextual \nEmbeddings for the incorrect MWEs class', fontsize=fontsize)

ax.set_ylabel('F1-score Gain \nfor the incorrect MWEs class', fontsize=fontsize)
ax.set_xlabel('Model', fontsize=fontsize)

plt.yticks(fontsize=fontsize)
plt.xticks(fontsize=fontsize)

ax.set(ylim=(0, 0.18))

plt.tight_layout()

plt.savefig('gain_incorrect_mwe.pdf', dpi=600)
```
